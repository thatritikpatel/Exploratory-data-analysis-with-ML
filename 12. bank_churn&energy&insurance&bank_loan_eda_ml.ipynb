# LOGISTIC REGRESSION DATASET (BANK CHURN)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('datasets\Bank_churn.csv')
df
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df['Geography'].unique()
```

```python
df.Gender.unique()
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'],axis = 1)
df
```

```python
sns.boxplot([df.Balance,df.EstimatedSalary])
```

```python
df.Exited.value_counts()
```

```python
#pip install imblearn
```

```python
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first = True,dtype = int)
df
```

```python
X = df.drop('Exited',axis =1)
Y= df.Exited
```

```python
from imblearn.over_sampling import SMOTE
model = SMOTE()
```

```python
x,y = model.fit_resample(X,Y)
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
                                            
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score , classification_report 
```

```python
print("Accuracy score is :",accuracy_score(ytest,y_pred))
print(classification_report(ytest,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

params = {'penalty' : ('l1','l2','ElasticNet'),'C':[1,2,3,4,6,10]}
clr = LogisticRegression()
clf = GridSearchCV(clr , param_grid = params,cv =5)
```

```python
clf
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
log_reg = LogisticRegression(penalty = 'l2' , C =1) 
log_reg.fit(xtrain,ytrain)
y_pred = log_reg.predict(xtest)
y_pred
```

```python
print(classification_report(ytest,y_pred))
```

```python
## We got a ACCURACY of 79%
```

```python
from sklearn.metrics import roc_auc_score
```

```python
print(roc_auc_score(ytest,y_pred))
```
