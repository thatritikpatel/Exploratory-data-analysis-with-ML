```python
X = df.drop('Alcohol',axis=1)
Y = df['Alcohol']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y ,test_size = 0.2,random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print("The accuracy of this model is :",accuracy_score(y_pred,Y_test))
print("\n")
print("The confusion matrix of this model is :",confusion_matrix(y_pred,Y_test))
```

```python
from sklearn.model_selection import GridSearchCV
param = {'penalty': ('l1', 'l2', 'elasticnet'), 'C': [1, 5, 9, 12], 'l1_ratio': [0.1, 0.5, 0.7]}
clf = GridSearchCV(model, param_grid=param, cv=5)
clf.fit(X_train, Y_train)
best_score = clf.best_score_
```

```python
print("By doing hyperparameter tunning is :",best_score)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
nb_clf = GaussianNB()
lr_clf = LogisticRegression()
dt_clf = DecisionTreeClassifier()
svm_clf = SVC(kernel = 'linear',probability = True)
```

```python
from sklearn.ensemble import VotingClassifier
estimator = [('decision_tree',dt_clf),('naive_bayes',nb_clf),('log_reg',lr_clf),('svm',svm_clf)]
ensemble_clf = VotingClassifier(estimators=estimator,voting='soft')
ensemble_clf.fit(X_train,Y_train)
y_ensemble_pred = ensemble_clf.predict(X_test)
```

```python
print("By Bagging the accuracy of this model is :",accuracy_score(y_ensemble_pred,Y_test))
```

```python
# By using Boosting method :->
from sklearn.ensemble import AdaBoostClassifier
classifier = AdaBoostClassifier()
classifier.fit(X_train,Y_train)
y_boosting_pred= classifier.predict(X_test)
print("The Accuracy score is by using Adaboostingmethod: ",accuracy_score(y_boosting_pred,Y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
classifier = GradientBoostingClassifier()
classifier.fit(X_train,Y_train)
y_boosting_pred= classifier.predict(X_test)
print("The Accuracy score is by using Gradiet Boosting method:",accuracy_score(y_boosting_pred,Y_test))
print("The confusion matrix is :\n",confusion_matrix(y_boosting_pred,Y_test))
```
