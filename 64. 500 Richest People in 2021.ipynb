```python
# Importing pandas and numpy for data preprocessing
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets\500 richest people 2021.csv', delimiter=';')
df
```

```python
df.drop(['Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10'], axis=1, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df = df.replace('Canda','Canada', regex=True)
```

```python
df
```

```python
df = df.replace('\$','', regex=True)
df[['Total Net Worth','$ Last Change']] = df[['Total Net Worth','$ Last Change']].replace(['B','M','K'],'', regex=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Total Net Worth'] = pd.to_numeric(df['Total Net Worth'],errors='coerce')
df['$ Last Change'] = pd.to_numeric(df['$ Last Change'],errors='coerce')
# df['$ YTD Change'] = pd.to_numeric(df['$ YTD Change'],errors='ignore')
```

```python
df.info()
```

```python
def value_to_float(x):
    if 'K' in x:
        return float(x.replace('K',''))*0.000001
    if 'M' in x:
        return float(x.replace('M',''))*1000
    if 'B' in x:
        return float(x.replace('B',''))

df['$ YTD Change'] = df['$ YTD Change'].apply(value_to_float)
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
from matplotlib import figure
style.use('fivethirtyeight')
```
