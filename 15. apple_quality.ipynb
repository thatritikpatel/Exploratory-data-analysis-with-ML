```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('datasets/apple_quality.csv')
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.Acidity.unique()
```

```python
df['Acidity'] = df.Acidity.astype('float')
```

```python
df.info()
```

```python
df.Quality.unique()
```

```python
df['Quality'] = df.Quality.map({'good':1, 'bad':0})
```

```python
df['Quality'].unique()
```

```python
df.Quality.value_counts()
```

```python
df
```

```python
df.drop('A_id', axis=1, inplace=True)
```

```python
sns.boxplot(df[df.columns])
plt.xticks(rotation=-40)
```

```python
Q1 = df.drop('Quality',axis=1).quantile(0.25)
Q3 = df.drop('Quality',axis=1).quantile(0.75)

IQR = Q3 - Q1

lower_fence = Q1 -1.5 *IQR
upper_fence = Q3 + 1.5 *IQR
```

```python
lower_fence
```

```python
upper_fence
```

```python
outlier_treated = {}
for i, col in enumerate(df.columns):
    if df.columns[i] == 'Quality':
        break
    df_outlier = np.where((df[col] >= upper_fence[i]) | (df[col] <= lower_fence[i]), df[col].mean(), df[col])
    outlier_treated[list(df.columns)[i]] = df_outlier
print(outlier_treated)
```

```python
outlier_treated.keys()
```

```python
df_final = pd.DataFrame(outlier_treated, columns=outlier_treated.keys())
```

```python
df_final['Quality'] = df.Quality
```

```python
df_final
```

```python
sns.boxplot(df_final[df_final.columns])
plt.xticks(rotation=40)
```

```python
lower_cap = df_final.drop(['Acidity', 'Quality'], axis=1).quantile(0.05)
upper_cap = df_final.drop(['Acidity', 'Quality'], axis=1).quantile(0.95)
```

```python
lower_cap
```

```python
upper_cap
```

```python
df_final['Size'] = np.where(df_final['Size'] < lower_cap['Size'], lower_cap['Size'],
        np.where(df_final['Size'] > upper_cap['Size'], upper_cap['Size'],
                df_final['Size']))
```

```python
df_final['Weight'] = np.where(df_final['Weight'] < lower_cap['Weight'], lower_cap['Weight'],
        np.where(df_final['Weight'] > upper_cap['Weight'], upper_cap['Weight'],
                df_final['Weight']))
df_final['Sweetness'] = np.where(df_final['Sweetness'] < lower_cap['Sweetness'], lower_cap['Sweetness'],
        np.where(df_final['Sweetness'] > upper_cap['Sweetness'], upper_cap['Sweetness'],
                df_final['Sweetness']))
df_final['Crunchiness'] = np.where(df_final['Crunchiness'] < lower_cap['Crunchiness'], lower_cap['Crunchiness'],
        np.where(df_final['Crunchiness'] > upper_cap['Crunchiness'], upper_cap['Crunchiness'],
                df_final['Crunchiness']))
df_final['Juiciness'] = np.where(df_final['Juiciness'] < lower_cap['Juiciness'], lower_cap['Juiciness'],
                                np.where(df_final['Juiciness'] > upper_cap['Juiciness'], upper_cap['Juiciness'],
                df_final['Juiciness']))
df_final['Ripeness'] = np.where(df_final['Ripeness'] < lower_cap['Ripeness'], lower_cap['Ripeness'],
                               np.where(df_final['Ripeness'] > upper_cap['Ripeness'], upper_cap['Ripeness'],
                df_final['Ripeness']))
```

```python
sns.boxplot(df_final[df_final.columns])
plt.xticks(rotation=40)
```

```python
# X and y split
x = df_final.drop('Quality', axis= 1)
y = df_final['Quality']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
param = {
    'C':[0.1,0.05,0.3,0.5,1.0],
    'penalty':['l1', 'l2', 'elasticnet'],
    'solver' : ['lbfgs','liblinear']
}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(LogisticRegression(), param_grid=param, cv=5, verbose=3, scoring='accuracy')
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
cls = LogisticRegression(C=0.1, penalty='l1', solver='liblinear')
```

```python
cls.fit(x_train, y_train)
```

```python
y_pred = cls.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
