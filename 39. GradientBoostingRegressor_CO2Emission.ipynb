```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("CO2_emission.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(['Make', 'Model', 'Vehicle Class'], axis=1, inplace=True)
```

```python
cat_col = [col for col in df.columns if df[col].dtype == 'O']
cat_col
```

```python
for col in cat_col:
    print(col, df[col].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cat_col:
    df[col] = le.fit_transform(df[col])
```

```python
df.info()
```

```python
df.head()
```

```python
X, y = df.drop('CO2 Emissions(g/km)', axis=1), df['CO2 Emissions(g/km)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor()
```

```python
model = model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

```python
param_grid = {
    'n_estimators': [100, 250, 400],
    'learning_rate': [0.01, 0.1, 0.05, 0.3]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='r2')
```

```python
grid = grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_grid = grid.predict(X_test)
```

```python
print(r2_score(y_test, y_pred_grid))
```
