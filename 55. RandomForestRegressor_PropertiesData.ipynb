```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("properties_data.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['neighborhood'].unique()
```

```python
df['quality'].unique()
```

```python
bool_col = [col for col in df.columns if df[col].dtype == 'bool']
bool_col
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in bool_col:
    df[col] = le.fit_transform(df[col])
```

```python
df.info()
```

```python
df['neighborhood'] = le.fit_transform(df['neighborhood'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Low', 'Medium','High', 'Ultra']])
df['quality'] = oe.fit_transform(df[['quality']])
```

```python
df.info()
```

```python
df.drop('price', axis=1, inplace=True)
```

```python
X, y = df.drop('price_per_sqft', axis=1), df['price_per_sqft']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

```python
param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [None, 10, 20, 30, 50],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_grid = grid.predict(X_test)
```

```python
print(r2_score(y_test, y_pred_grid))
```
