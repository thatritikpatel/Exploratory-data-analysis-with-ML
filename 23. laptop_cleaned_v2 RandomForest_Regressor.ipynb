```python
x=df.drop('price',axis=1)
y=df['price']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
```

```python
from sklearn.linear_model import  LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
```

```python
p1=Pipeline([('lg',LinearRegression())])
p2=Pipeline([('dtr',DecisionTreeRegressor())])
p3=Pipeline([('svr',SVR())])
p4=Pipeline([('rfr',RandomForestRegressor())])
```

```python
pipelines=[p1,p2,p3,p4]
for pipe in pipelines:
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(oob_score=True)
rfr
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'max_depth':[1,50,100,150,200],'n_estimators':[50,100,200]}
grid_search=GridSearchCV(rfr,param_grid=param_grid,cv=5,verbose=3)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_model=grid_search.best_estimator_
best_model
```

```python
y_pred_tuned=best_model.predict(x_test)
y_pred_tuned
```

```python
r2_score(y_test,y_pred_tuned)
```
