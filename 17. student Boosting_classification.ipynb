```python
cols=['High_School_Type','Attendance','Grade']
from sklearn.preprocessing import LabelEncoder
for col in df.select_dtypes(include=['object']).columns:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
x=df.drop('Project_work',axis=1)
y=df['Project_work']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_test.shape,x_train.shape
```

```python
#AdaBoosting classifier
from sklearn.ensemble import AdaBoostClassifier
abc=AdaBoostClassifier()
abc.fit(x_train,y_train)
y_pred=abc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Hyperparametre tuning for adaboosting 
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[50,100,200],'learning_rate':[0.01,0.1,1.0,1.5,2.0],'algorithm':['SAMME','SAMMER']}
grid_search=GridSearchCV(estimator=abc,param_grid=param_grid,cv=5,verbose=3)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_model=grid_search.best_estimator_
y_pred_tuned=best_model.predict(x_test)
y_pred_tuned
```

```python
accuracy_score(y_test,y_pred_tuned)
```

```python
print(classification_report(y_test,y_pred_tuned))
```

```python
#Gradient Boosting classifier
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
gbc.fit(x_train,y_train)
y_pred=gbc.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Hyperparametre tuning for gradient boosting 
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[50,100,200],'learning_rate':[0.01,0.1,0.05,0.2]}
grid_search=GridSearchCV(estimator=gbc,param_grid=param_grid,cv=5,verbose=3)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_model=grid_search.best_estimator_
y_pred_tuned=best_model.predict(x_test)
y_pred_tuned
```

```python
accuracy_score(y_test,y_pred_tuned)
```

```python
print(classification_report(y_test,y_pred_tuned))
```
